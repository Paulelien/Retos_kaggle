```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
remote_set = pd.read_csv(r"C:\Users\ricar\OneDrive\Escritorio\DS\Retos Kaggle\Impact_of_Remote_Work_on_Mental_Health.csv")
remote_set.head (5)
```

```python
sns.boxplot(x="Productivity_Change", y="Hours_Worked_Per_Week", data=remote_set)
plt.title("Distribución de Variable Continua por Categorías")
plt.show()
```

```python
# Análisis univariado (distribución de variables numéricas)

for col in remote_set.select_dtypes(include='number').columns:
    sns.histplot(remote_set[col], kde=True)
    plt.title(f"Distribución de {col}")
    plt.show()
```
